```
import pandas as pd #for tables
import numpy as np #for maths
import seaborn as sns  #graphs
import matplotlib.pyplot as plt #graphs
%matplotlib inline #to make sure graph is in collab
```

```
from sklearn.datasets import make_blobs #generating random dataset 
```

```
data = make_blobs(n_samples = 500, n_features = 2, centers = 6, cluster_std = 2.5, random_state = 201)  #sample size 500,6 clusters, standard dev = 2.5, random state output generated will be same 
```

```
data[0].shape
```

```
plt.scatter(data[0][:,0],data[0][:,1], c=data[1]) #for plotting c is color sequence
```

```
from sklearn.cluster import KMeans #importing k-means from clustering algorithms
```

```
kmeans = KMeans(n_clusters = 4) #considering 4 clusters
kmeans.fit(data[0])  #fitting the data in 4 clustering
```

```
kmeans.cluster_centers_
```

```
plt.scatter(kmeans.cluster_centers_[:,0],kmeans.cluster_centers_[:,1])
```

```
kmeans.labels_ #no of clusters 
```

```
data[1] #cluster number 
```

```
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap = 'rainbow')
```

```
kmeans2 = KMeans(n_clusters = 5)
kmeans2.fit(data[0])
```

```
fig, (ax1,ax2) = plt.subplots(1,2,sharey=True, figsize=(10,6))

ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1], c=kmeans2.labels_, cmap='rainbow')

ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1], c=data[1], cmap = 'rainbow')
```
